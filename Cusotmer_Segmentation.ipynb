## clustering

```python

# cluster on 2 features (annual income , spending score )
df1 = data_set[['Annual Income (k$)', 'Spending Score (1-100)']]
df1.head()
```

```python
# scatter plot
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', data=df1)
```

```python
from sklearn.cluster import KMeans
errors = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(df1)
    errors.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(range(1, 11), errors, marker='o')
plt.title('Elbow Method For Optimal k')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python

km = KMeans(n_clusters=5)
km.fit(df1)
y = km.predict(df1)
df1['Label'] = y
df1.head()
```

```python
X = df2[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]
Y = df2['Label']
```

```python
print(X)
```

```python
print(Y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train, Y_train)
```

#### Building a Predictive Function

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
test_data_prediction = model.predict(X_test)
```

```python
print(test_data_prediction)
```

```python
import joblib
joblib.dump(model, 'DecisionTreeClassifier_model.pkl')
```

```python
from sklearn.metrics import accuracy_score
test_data_accuracy = accuracy_score(Y_test, test_data_prediction)
print(test_data_accuracy)
```
